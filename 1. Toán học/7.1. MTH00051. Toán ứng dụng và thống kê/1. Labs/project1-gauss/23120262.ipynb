## Đề bài

1. **Sinh viên viết hàm `Gauss_elimination(A)`**, trong đó:
   - **Input:** `A`, là ma trận mở rộng của hệ phương trình \(Ax = b\).
   - **Output:** ma trận có dạng bậc thang có được từ ma trận `A`.
   - **Lưu ý:** Sinh viên không được dùng các hàm có sẵn của các thư viện để tìm ma trận bậc thang.

2. **Sinh viên viết hàm `back_substitution(A)`**, trong đó:
   - **Input:** `A`, là ma trận có dạng bậc thang thu được từ ma trận mở rộng của hệ phương trình \(Ax = b\).
   - **Output:** nghiệm của hệ phương trình (trường hợp nghiệm duy nhất/vô số nghiệm) hoặc thông báo hệ phương trình vô nghiệm.
   ## Ý Tưởng Thực Hiện

   ### Thuật Toán Khử Gauss

   #### Phần 1: Đưa Ma Trận Về Dạng Bậc Thang

   1. **Bước 1:** Xác định cột trái nhất không chứa toàn số 0.
   2. **Bước 2:** Đổi chỗ hai dòng, nếu cần thiết, để đưa số hạng khác 0 nào đó ở dưới về đầu cột nhận được ở Bước 1.
   3. **Bước 3:** Với số hạng đầu cột nhận được từ Bước 2, nhân dòng chứa nó để có số dẫn đầu 1.
   4. **Bước 4:** Cộng một bội số thích hợp của dòng đầu cho từng dòng dưới để biến các số hạng bên dưới số dẫn đầu thành 0.
   5. **Bước 5:** Che dòng đầu đã làm xong. Lặp lại các bước cho đến khi được ma trận bậc thang.

   #### Phần 2: Thế Ngược

   1. **Bước 1:** Bắt đầu từ dòng cuối, ta thế ngược lại hệ số vào dòng cuối của hệ phương trình, rồi giải nghiệm tương ứng.
   2. **Bước 2:** Từ dòng cuối cùng lên trên, ta lấy nghiệm vừa tìm được thế vào phương trình. Vậy là ta sẽ tìm được nghiệm thứ 2. Cứ như vậy cho đến khi tìm được hết các nghiệm.

   #### Lưu ý

   - Nếu có trường hợp \( 0x = a \) với \( a != 0 \) thì kết luận phương trình vô nghiệm.
   - Nếu có trường hợp \( 0x = a \) với \( a = 0 \) hoặc hệ phương trình có \( n \) nghiệm nhưng chỉ có \( m \) dòng với \( m < n \) thì kết luận phương trình vô số nghiệm.

   ## Mô Tả Các Hàm

   ### Hàm `Gauss_elimination(A)`
   - **Dữ liệu truyền vào:** Ma trận bổ sung (augmented matrix) \( A \).
   - **Mục đích:** Đưa ma trận về dạng bậc thang.
   - **Cách hoạt động:**
      - Đầu tiên, xác định số hàng và số cột của ma trận \( A \).
      - Ở vòng lặp ngoài, duyệt từ hàng đầu tiên đến hàng áp chót.
      - Ở vòng lặp trong, duyệt các hàng dưới hàng hiện tại ở vòng lặp ngoài.
      - Nếu phần tử tại vị trí \( (i, k) \) khác 0, tính giá trị \( lambda \) bằng cách lấy phần tử đầu tiên khác 0 ở hàng bên dưới hàng đang xét, chia cho phần tử đầu tiên ở hàng trên.
      - Sử dụng \( lambda \) để đưa các phần tử bên dưới phần tử cơ sở về 0.
      - Lặp lại quá trình cho đến khi thu được ma trận bậc thang.

   ### Hàm `back_substitution(A)`

   - **Dữ liệu truyền vào:** Ma trận bậc thang \( A \).
   - **Mục đích:** Giải hệ phương trình từ ma trận bậc thang.
   - **Cách hoạt động:**
      - Đầu tiên, xác định số hàng và số cột của ma trận \( A \).
      - Xét các trường hợp vô nghiệm và vô số nghiệm.
      - Nếu không có trường hợp đặc biệt, tạo một vector để chứa kết quả của các nghiệm.
      - Tính giá trị của biến cuối, rồi tiếp tục thế lên phương trình phía trên các giá trị đã biết để tìm giá trị chưa biết.
      - Cuối cùng, in ra kết quả.


```python
def Gauss_elimination(A):
    row = len(A)
    col = len(A[0])
    for k in range(row - 1):
        if A[k][k] == 0:
            for i in range(k + 1, row):
                if A[i][k] != 0:
                    A[k], A[i] = A[i], A[k]
                    break
            else:
                continue
        pivot = A[k][k]
        for i in range(k + 1, row):
            if A[i][k] != 0:
                lam = A[i][k] / pivot
                A[i][k:] = [A[i][j] - lam * A[k][j] for j in range(k, col)]

def back_substitution(A):
    row = len(A)
    col = len(A[0])

    # Kiểm tra hệ phương trình vô nghiệm
    for i in range(row):
        if all(A[i][j] == 0 for j in range(col - 1)) and A[i][col - 1] != 0:
            print("He phuong trinh vo nghiem")
            for k in range(col - 1):
                print(f"x[{k+1}] = ?")
            return

    # Kiểm tra hệ phương trình vô số nghiệm
    if row < col - 1 or any(all(A[i][j] == 0 for j in range(col - 1)) and A[i][col - 1] == 0 for i in range(row)):
        print("He p huong trinh vo so nghiem")
        
        num_free_vars = col - 1 - row
        free_vars = [f"a{i+1}" for i in range(num_free_vars)]
        
        solutions = {}
        for i in range(row - 1, -1, -1):
            pivot_col = -1
            for j in range(col - 1):
                if A[i][j] != 0:
                    pivot_col = j
                    break
            if pivot_col == -1:
                continue
            
            expr = f"{A[i][col - 1]}"
            for j in range(pivot_col + 1, col - 1):
                if A[i][j] != 0:
                    if j >= row:
                        var_name = free_vars[j - row]
                    else:
                        var_name = solutions.get(j, f"x[{j+1}]")
                    if A[i][j] == 1:
                        expr += f" + {var_name}"
                    elif A[i][j] == -1:
                        expr += f" - {var_name}"
                    else:
                        expr += f" + {A[i][j]} * {var_name}" if A[i][j] > 0 else f" - {-A[i][j]} * {var_name}"
            divisor = A[i][pivot_col]
            if divisor == 1:
                solutions[pivot_col] = expr
            else:
                solutions[pivot_col] = f"({expr}) / {divisor}"
        
        free_val = 1
        for i in range(col - 1):
            if i in solutions:
                print(f"x[{i+1}] = {solutions[i]}")
            else:
                if i >= row:
                    print(f"x[{i+1}] = {free_vars[i - row]}")
                else:
                    print(f"x[{i+1}] = a{free_val}, ∀ a{free_val} ∈ R")
                    free_val += 1
        return

    # Tính nghiệm duy nhất
    x = [0] * (col - 1)
    for i in range(row - 1, -1, -1):
        x[i] = (A[i][col - 1] - sum(A[i][j] * x[j] for j in range(i + 1, col - 1))) / A[i][i]
    print("Nghiem he pt lan luot la: ", x)

def print_matrix(A):
    for row in A:
        print("[", " ".join(f"{elem:5.1f}" for elem in row), "]")

A = [
    [1, 2, 0, 2, 6],
    [3, 5, -1, 6, 17],
    [2, 4, 1, 2, 12],
    [2, 0, -7, 11, 7]
]

# A = [
#     [1, -2, -1, 1],
#     [2, -3, 1, 6],
#     [3, -5, 0, 7],
#     [1, 0, 5, 9]
# ]

print("\nMa tran ban dau: ")
print_matrix(A)
Gauss_elimination(A)

print("\nMa tran bac thang: ")
print_matrix(A)

print("\nNghiem: ")
back_substitution(A)
```
